```python
import dash
#from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
from pyngrok import ngrok

d=[]
for i in range(30):
  tmp=[]
  for month in range(12):
    dd=0
    tt=0
    for yr in range(10):
      tt+=total_flights[i][month][yr]
      dd+=delay_flights[i][month][yr]
    tmp.append(dd/tt*100)
  d.append(tmp)

d_m=[]
for i in range(30):
  tmp=[]
  for month in range(12):
    mm=0
    dd=0
    for yr in range(10):
      dd+=delay_flights[i][month][yr]
      mm+=delay_minutes[i][month][yr]
    tmp.append(mm/dd)
  d_m.append(tmp)

from sklearn.preprocessing import MinMaxScaler
import numpy as np

airport_rating = []
for i in range(30):
  tmp=0
  for j in range(12):
    tmp+=d[i][j]*d_m[i][j]
  airport_rating.append(tmp)
airport_rating = np.array(airport_rating).reshape(-1, 1)

scaler = MinMaxScaler()
normalized_lst = scaler.fit_transform(airport_rating)

airport_rating = normalized_lst.flatten()



d_c=[]
for i in range(39):
  tmp=[]
  for month in range(12):
    dd=0
    tt=0
    for yr in range(10):
      tt+=total_flights_cr[i][month][yr]
      dd+=delay_flights_cr[i][month][yr]
    if tt==0:
      tmp.append(0)
    else:
      tmp.append(dd/tt*100)
  d_c.append(tmp)

d_c_m=[]
for i in range(39):
  tmp=[]
  for month in range(12):
    mm=0
    dd=0
    for yr in range(10):
      dd+=delay_flights_cr[i][month][yr]
      mm+=delay_minutes_cr[i][month][yr]
    if dd==0:
      tmp.append(0)
    else:
      tmp.append(mm/dd)
  d_c_m.append(tmp)

carrier_rating = []
for i in range(39):
  tmp=0
  for j in range(12):
    tmp+=d_c[i][j]*d_c_m[i][j]
  carrier_rating.append(tmp)
carrier_rating = np.array(carrier_rating).reshape(-1, 1)

scaler = MinMaxScaler()
normalized_lst = scaler.fit_transform(carrier_rating)

carrier_rating = normalized_lst.flatten()
carrier_rating[9] = 0.01

dcarrier = pd.DataFrame({
    'y': [i for i in range(39)],
    'carrier': all_carrier,
    'index': carrier_rating,
})
fig_carrier = px.bar(
    dcarrier,
    x = 'carrier',
    y = 'index',
    hover_name='carrier',
    hover_data={'index': True},
    color='index',
    title="All Carriers"
)

data = pd.DataFrame({
    'latitude': lat,
    'longitude': lon,
    'location': loc,
    'index': airport_rating,
})

fig_map = px.scatter_mapbox(
    data,
    lat='latitude',
    lon='longitude',
    hover_name='location',
    hover_data={'index': True},
    color='index',
    size='index',
    title="Major Airports"
)

fig_map.update_layout(
    mapbox_style="carto-positron",
    mapbox_zoom=2,
    mapbox_center={"lat": 37.0902, "lon": -95.7129}
)

fig_bar = px.bar(
    title="fig 1: Delayed Percentage by Airport"
)
fig_bar.update_layout(
    xaxis_title='month',
    yaxis_title='delayed percentage (%)',
    xaxis=dict(range=[1, 12]),
    yaxis=dict(range=[0, 100])
)

fig_pie = px.pie(pd.DataFrame({'Category': [], 'Value': []}), names='Category', values='Value', title="fig 2: Delayed Causes Proportion")

fig_bar2 = px.bar(
    title="fig 3: Delayed Minutes per Delay Occurance"
)
fig_bar2.update_layout(
    xaxis_title='month',
    yaxis_title='minutes',
    xaxis=dict(range=[1, 12]),
)

fig_bar3 = px.bar(
    title="fig 4: Delayed Minutes by Causes"
)
fig_bar3.update_layout(
    xaxis_title='Causes',
    yaxis_title='minutes',
    xaxis=dict(range=[1, 5]),
)


app = dash.Dash(__name__)

MONTH = ['Jan.', 'Feb.', 'Mar.', 'Apr.', 'May', 'Jun.', 'Jul.', 'Aug.', 'Sep.', 'Oct.', 'Nov.', 'Dec.']
causes = ['Carrier', 'Weather', 'NAS', 'Security', 'Late Aircraft']

app.layout = html.Div([
    #section 1
    html.P("In this project, we utilize data from the Bureau of Transportation Statistics to gain valuable insights into U.S. flight patterns. Our analysis focuses on identifying which airports and airlines experience the most delays, as well as pinpointing the most common causes of these delays. By understanding these factors, we aim to provide travelers information that can help them minimize the likelihood of delays."),
    html.H1("Delayed Percentage for Major Airports"),
    html.P("In this section, map with major airports in US are displayed. Each airport has an index, the value of index implies the expected minutes passengers will have to wait per flight in this airport."),
    html.Br(),
    html.P("Hover through these airports to gain insights including the ratio of flights that are delayed(fig 1), ratio of delayed causes(fig 2), average delayed minutes for a delayed flight(fig 3), and average delayed minutes for specific delayed cause(fig 4)."),

    html.Div([
          dcc.Graph(id="map-graph", figure=fig_map)
      ], style={'grid-column': '2'}),
    html.P("Hover through individual months to check the ratio of causes."),
    html.Div([
        html.Div([
            dcc.Graph(id="bar-graph", figure=fig_bar)
        ], style={'flex': 1, 'padding': '10px'}),
        html.P("Change the Result by Clicking Fig 1."),
        html.Div([
            dcc.Graph(id="pie-graph", figure=fig_pie)
        ], style={'flex': 1, 'padding': '10px'}),
        html.Div([
            dcc.Graph(id="bar-graph-2", figure=fig_bar2)
        ], style={'flex': 1, 'padding': '10px'}),
        html.P("Change the Result by Clicking Fig 1."),
        html.Div([
            dcc.Graph(id="bar-graph-3", figure=fig_bar3)
        ], style={'flex': 1, 'padding': '10px'}),
    ], style={'display': 'flex',
        'flex-wrap': 'wrap',
        'gap': '10px',
        'justify-content': 'space-between',
        }),


    #section 2
    html.H1("Delayed Percentage for Carriers"),
    html.P("In this section, all the carriers are displayed and the height of the bar represents the expected minutes passengers will have to wait per flight when choosing the specific carrier."),
    html.Br(),
    html.P("Hover through these carriers to gain insights including the ratio of flights that are delayed(fig 1), ratio of delayed causes(fig 2), average delayed minutes for a delayed flight(fig 3), and average delayed minutes for specific delayed cause(fig 4)."),
    html.Div([
        dcc.Graph(id="carrier-graph", figure=fig_carrier)
    ], style={'grid-column': '2'}),
    html.P("Hover through individual months to check the ratio of causes."),
    html.Div([
        html.Div([
            dcc.Graph(id="carrier-bar-graph", figure=fig_bar_cr)
        ], style={'flex': 1,}),
        html.P("Change the Result by Clicking Fig 1."),
        html.Div([
            dcc.Graph(id="carrier-pie-graph", figure=fig_pie_cr)
        ], style={'flex': 1,}),
        html.Div([
            dcc.Graph(id="carrier-bar-graph-2", figure=fig_bar2_cr)
        ], style={'flex': 1,}),
        html.P("Change the Result by Clicking Fig 1."),
        html.Div([
            dcc.Graph(id="carrier-bar-graph-3", figure=fig_bar3_cr)
        ], style={'flex': 1,}),
    ], style={'display': 'flex',
    'flex-wrap': 'wrap',
    'justify-content': 'space-between',
    'gap': '1px',}),

    #section 3
    html.H1("Carriers Performance by Airport and Month"),
    html.P("In this section, pick your airport and check the on-time performance for each carrier to choose the flight with the least chance of experiencing a delay."),
    html.Div([
      dcc.Graph(id="map-graph-2", figure=fig_map)
    ], style={'flex': 1, 'width': '100%'}),
    dcc.Checklist(
        id='category-checklist',
        options=[
            {'label': MONTH[i], 'value': MONTH[i]} for i in range(12)
        ],
        value=[],
        inline=True
    ),

    dcc.Graph(id='search-graph'),

    #section 4
    html.H1("Delayed Causes"),
    html.P("In this section, select delay causes to gain insights about what is the most common cause of flight delays, and how long passengers have to wait on average."),

    dcc.Dropdown(
        id='category-dropdown',
        options=[{'label': cause, 'value': cause} for cause in causes],
        value='Carrier',  # default category
        style={'width': '50%'}
    ),

    dcc.Graph(id='causes-graph'),
])

# CALLBACK FOR AIRPORT 1
@app.callback(
    Output("bar-graph", "figure"),
    Input("map-graph", "hoverData")
)
def update_airport_bar_chart(hoverData):
    if hoverData is None:
      return fig_bar

    hovered_airport = hoverData['points'][0]['hovertext']

    data = pd.DataFrame({
        'location': [hovered_airport for i in range(12)],
        'delay': d[airport_to_num[hovered_airport]],
        'month': [i for i in range(1,13)]
    })

    fig = px.bar(
        data_frame=data,
        x="month",
        y="delay",
        hover_name='location',
        hover_data={'month': True},
        title=f"fig 1: Delayed Percentage at<br>{hovered_airport}"
    )
    fig.update_traces(
        customdata=data[['month']].values
    )
    return fig

# CALLBACK FOR AIRPORT 2
@app.callback(
    Output("pie-graph", "figure"),
    Input("bar-graph", "hoverData")
)
def update_airport_pie_chart(hoverData):
    if hoverData is None:
        return fig_pie

    hovered_airport = hoverData['points'][0]['hovertext']
    mm = hoverData['points'][0]['customdata'][0]

    data = pd.DataFrame({
        'causes': ['Carrier','Weather','NAS','Security','Late Aircraft'],
        'prop': [sum(delay_carrier[airport_to_num[hovered_airport]][mm-1])/sum(delay_flights[airport_to_num[hovered_airport]][mm-1])*100,
                 sum(delay_weather[airport_to_num[hovered_airport]][mm-1])/sum(delay_flights[airport_to_num[hovered_airport]][mm-1])*100,
                 sum(delay_nas[airport_to_num[hovered_airport]][mm-1])/sum(delay_flights[airport_to_num[hovered_airport]][mm-1])*100,
                 sum(delay_security[airport_to_num[hovered_airport]][mm-1])/sum(delay_flights[airport_to_num[hovered_airport]][mm-1])*100,
                 sum(delay_late[airport_to_num[hovered_airport]][mm-1])/sum(delay_flights[airport_to_num[hovered_airport]][mm-1])*100,
                 ],
    })

    #MONTH = ['Jan.', 'Feb.', 'Mar.', 'Apr.', 'May', 'Jun.', 'Jul.', 'Aug.', 'Sep.', 'Oct.', 'Nov.', 'Dec.']
    fig = px.pie(
        data,
        names="causes",
        values="prop",
        color="causes",
        color_discrete_sequence=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"],
        title=f"fig 2: Delayed Causes at<br>{hovered_airport}, {MONTH[mm-1]}"
    )
    return fig

# CALLBACK FOR AIRPORT 3
@app.callback(
    Output("bar-graph-2", "figure"),
    Input("map-graph", "hoverData")
)
def update_airport_bar_chart_minutes(hoverData):
    if hoverData is None:
      return fig_bar2

    hovered_airport = hoverData['points'][0]['hovertext']
    data = pd.DataFrame({
        'location': [hovered_airport for i in range(12)],
        'delay': d_m[airport_to_num[hovered_airport]],
        'month': [i for i in range(1,13)]
    })

    fig = px.bar(
        data_frame=data,
        x="month",
        y="delay",
        hover_name='location',
        hover_data={'month': True},
        title=f"fig 3: Delayed Minutes at<br>{hovered_airport}"
    )
    fig.update_traces(
        customdata=data[['month']].values
    )
    return fig

# CALLBACK FOR AIRPORT 4
@app.callback(
    Output("bar-graph-3", "figure"),
    Input("bar-graph", "hoverData")
)
def update_airport_bar_chart_minutes(hoverData):
    if hoverData is None:
        return fig_bar3

    hovered_airport = hoverData['points'][0]['hovertext']
    mm = hoverData['points'][0]['customdata'][0]
    data = pd.DataFrame({
      'minutes': [sum(delay_minutes_carrier[airport_to_num[hovered_airport]][mm-1])/sum(delay_carrier[airport_to_num[hovered_airport]][mm-1]),
                  sum(delay_minutes_weather[airport_to_num[hovered_airport]][mm-1])/sum(delay_weather[airport_to_num[hovered_airport]][mm-1]),
                  sum(delay_minutes_nas[airport_to_num[hovered_airport]][mm-1])/sum(delay_nas[airport_to_num[hovered_airport]][mm-1]),
                  sum(delay_minutes_security[airport_to_num[hovered_airport]][mm-1])/sum(delay_security[airport_to_num[hovered_airport]][mm-1]),
                  sum(delay_minutes_late[airport_to_num[hovered_airport]][mm-1])/sum(delay_late[airport_to_num[hovered_airport]][mm-1]),
                ],
      'causes': ['carrier', 'weather', 'nas', 'securtiy', 'late aircraft']
    })

    fig = px.bar(
      data_frame=data,
      x="causes",
      y="minutes",
      color="causes",
      color_discrete_sequence=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"],
      title=f"fig 4: Delayed Minutes by Causes",
    )
    return fig

# CALLBACK FOR CARRIER 1
@app.callback(
    Output("carrier-bar-graph", "figure"),
    Input("carrier-graph", "hoverData")
)
def update_carrier_bar_chart(hoverData):
    if hoverData is None:
        return fig_bar_cr
    hovered_carrier = hoverData['points'][0]['hovertext']

    data = pd.DataFrame({
        'location': [hovered_carrier for i in range(12)],
        'delay': d_c[carrier_to_num[hovered_carrier]],
        'month': [i for i in range(1,13)]
    })

    fig = px.bar(
        data_frame=data,
        x="month",
        y="delay",
        hover_name='location',
        hover_data={'month': True},
        title=f"fig 1: Delayed Percentage for<br>{hovered_carrier}"
    )
    fig.update_traces(
        customdata=data[['month']].values
    )
    return fig

# CALLBACK FOR CARRIER 2
@app.callback(
    Output("carrier-pie-graph", "figure"),
    Input("carrier-bar-graph", "hoverData")
)
def update_carrier_pie_chart(hoverData):
    if hoverData is None:
        return fig_pie_cr

    hovered_carrier = hoverData['points'][0]['hovertext']
    mm = hoverData['points'][0]['customdata'][0]

    data = pd.DataFrame({
        'causes': ['Carrier','Weather','NAS','Security','Late Aircraft'],
        'prop': [sum(delay_carrier_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_flights_cr[carrier_to_num[hovered_carrier]][mm-1])*100,
                 sum(delay_weather_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_flights_cr[carrier_to_num[hovered_carrier]][mm-1])*100,
                 sum(delay_nas_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_flights_cr[carrier_to_num[hovered_carrier]][mm-1])*100,
                 sum(delay_security_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_flights_cr[carrier_to_num[hovered_carrier]][mm-1])*100,
                 sum(delay_late_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_flights_cr[carrier_to_num[hovered_carrier]][mm-1])*100,
                 ],
    })

    MONTH = ['Jan.', 'Feb.', 'Mar.', 'Apr.', 'May', 'Jun.', 'Jul.', 'Aug.', 'Sep.', 'Oct.', 'Nov.', 'Dec.']
    fig = px.pie(
        data,
        names="causes",
        values="prop",
        color="causes",
        color_discrete_sequence=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"],

        title=f"fig 2: Delayed Causes for<br>{hovered_carrier}, {MONTH[mm-1]}"
    )
    return fig

# CALLBACK FOR CARRIER 3
@app.callback(
    Output("carrier-bar-graph-2", "figure"),
    Input("carrier-graph", "hoverData")
)
def update_carrier_bar_chart_minutes(hoverData):
    if hoverData is None:
      return fig_bar2_cr

    hovered_carrier = hoverData['points'][0]['hovertext']
    data = pd.DataFrame({
        'location': [hovered_carrier for i in range(12)],
        'delay': d_c_m[carrier_to_num[hovered_carrier]],
        'month': [i for i in range(1,13)]
    })

    fig = px.bar(
        data_frame=data,
        x="month",
        y="delay",
        hover_name='location',
        hover_data={'month': True},
        title=f"fig 3: Delayed Minutes for<br>{hovered_carrier}"
    )
    fig.update_traces(
        customdata=data[['month']].values
    )
    return fig

# CALLBACK FOR CARRIER 4
@app.callback(
    Output("carrier-bar-graph-3", "figure"),
    Input("carrier-bar-graph", "hoverData")
)
def update_carrier_bar_chart_minutes(hoverData):
    if hoverData is None:
        return fig_bar3_cr

    hovered_carrier = hoverData['points'][0]['hovertext']
    mm = hoverData['points'][0]['customdata'][0]
    data = pd.DataFrame({
      'minutes': [sum(delay_minutes_carrier_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_carrier_cr[carrier_to_num[hovered_carrier]][mm-1]),
                  sum(delay_minutes_weather_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_weather_cr[carrier_to_num[hovered_carrier]][mm-1]),
                  sum(delay_minutes_nas_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_nas_cr[carrier_to_num[hovered_carrier]][mm-1]),
                  sum(delay_minutes_security_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_security_cr[carrier_to_num[hovered_carrier]][mm-1]),
                  sum(delay_minutes_late_cr[carrier_to_num[hovered_carrier]][mm-1])/sum(delay_late_cr[carrier_to_num[hovered_carrier]][mm-1]),
                ],
      'causes': ['carrier', 'weather', 'nas', 'securtiy', 'late aircraft']
    })

    fig = px.bar(
      data_frame=data,
      x="causes",
      y="minutes",
      color="causes",
      color_discrete_sequence=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"],
      title=f"fig 4: Delayed Minutes by Causes",
    )
    return fig


# CALLBACK FOR SEARCH
@app.callback(
    Output('search-graph', 'figure'),
    [Input('category-checklist', 'value'), Input('map-graph-2', 'clickData')]
)
def update_graph(selected_values, clickData):
    if clickData:
      clicked_airport = clickData['points'][0]['hovertext']
      data = {
        "Carriers": all_carrier,
        "Jan.": [delay_minutes_se[airport_to_num[clicked_airport]][0][i] / total_flights_se[airport_to_num[clicked_airport]][0][i] if total_flights_se[airport_to_num[clicked_airport]][0][i] != 0 else 0 for i in range(39)],
        "Feb.": [delay_minutes_se[airport_to_num[clicked_airport]][1][i] / total_flights_se[airport_to_num[clicked_airport]][1][i] if total_flights_se[airport_to_num[clicked_airport]][1][i] != 0 else 0 for i in range(39)],
        "Mar.": [delay_minutes_se[airport_to_num[clicked_airport]][2][i] / total_flights_se[airport_to_num[clicked_airport]][2][i] if total_flights_se[airport_to_num[clicked_airport]][2][i] != 0 else 0 for i in range(39)],
        "Apr.": [delay_minutes_se[airport_to_num[clicked_airport]][3][i] / total_flights_se[airport_to_num[clicked_airport]][3][i] if total_flights_se[airport_to_num[clicked_airport]][3][i] != 0 else 0 for i in range(39)],
        "May": [delay_minutes_se[airport_to_num[clicked_airport]][4][i] / total_flights_se[airport_to_num[clicked_airport]][4][i] if total_flights_se[airport_to_num[clicked_airport]][4][i] != 0 else 0 for i in range(39)],
        "Jun.": [delay_minutes_se[airport_to_num[clicked_airport]][5][i] / total_flights_se[airport_to_num[clicked_airport]][5][i] if total_flights_se[airport_to_num[clicked_airport]][5][i] != 0 else 0 for i in range(39)],
        "Jul.": [delay_minutes_se[airport_to_num[clicked_airport]][6][i] / total_flights_se[airport_to_num[clicked_airport]][6][i] if total_flights_se[airport_to_num[clicked_airport]][6][i] != 0 else 0 for i in range(39)],
        "Aug.": [delay_minutes_se[airport_to_num[clicked_airport]][7][i] / total_flights_se[airport_to_num[clicked_airport]][7][i] if total_flights_se[airport_to_num[clicked_airport]][7][i] != 0 else 0 for i in range(39)],
        "Sep.": [delay_minutes_se[airport_to_num[clicked_airport]][8][i] / total_flights_se[airport_to_num[clicked_airport]][8][i] if total_flights_se[airport_to_num[clicked_airport]][8][i] != 0 else 0 for i in range(39)],
        "Oct.": [delay_minutes_se[airport_to_num[clicked_airport]][9][i] / total_flights_se[airport_to_num[clicked_airport]][9][i] if total_flights_se[airport_to_num[clicked_airport]][9][i] != 0 else 0 for i in range(39)],
        "Nov.": [delay_minutes_se[airport_to_num[clicked_airport]][10][i] / total_flights_se[airport_to_num[clicked_airport]][10][i] if total_flights_se[airport_to_num[clicked_airport]][10][i] != 0 else 0 for i in range(39)],
        "Dec.": [delay_minutes_se[airport_to_num[clicked_airport]][11][i] / total_flights_se[airport_to_num[clicked_airport]][11][i] if total_flights_se[airport_to_num[clicked_airport]][11][i] != 0 else 0 for i in range(39)],
      }

      df = pd.DataFrame(data)
      fig = px.bar(df, x='Carriers', y=selected_values, title=f"Expected Delayed Time at {clicked_airport} Airport",barmode='group')
      fig.update_layout(yaxis_title="Minutes")
      return fig

    df_empty = pd.DataFrame({'Category': [], 'Minutes': []})
    fig = px.bar(df_empty, x='Category', y='Minutes', title="")
    return fig

data = {
    'month': MONTH,
    'Carrier percentages': [delay_ca[0][i]/delay_flights_ca[i]*100 for i in range(12)],
    'Carrier minutes': [delay_minutes_ca[0][i]/delay_ca[0][i] for i in range(12)],
    'Weather percentages': [delay_ca[1][i]/delay_flights_ca[i]*100 for i in range(12)],
    'Weather minutes': [delay_minutes_ca[1][i]/delay_ca[1][i] for i in range(12)],
    'NAS percentages': [delay_ca[2][i]/delay_flights_ca[i]*100 for i in range(12)],
    'NAS minutes': [delay_minutes_ca[2][i]/delay_ca[2][i] for i in range(12)],
    'Security percentages': [delay_ca[3][i]/delay_flights_ca[i]*100 for i in range(12)],
    'Security minutes': [delay_minutes_ca[3][i]/delay_ca[3][i] for i in range(12)],
    'Late Aircraft percentages': [delay_ca[4][i]/delay_flights_ca[i]*100 for i in range(12)],
    'Late Aircraft minutes': [delay_minutes_ca[4][i]/delay_ca[4][i] for i in range(12)],
}

dcauses = pd.DataFrame(data)
import plotly.graph_objects as go

# CALLBACK FOR CAUSES
@app.callback(
    Output('causes-graph', 'figure'),
    Input('category-dropdown', 'value')
)
def update_graph(selected_cause):
    mm = dcauses['month']
    y1_vals = dcauses[f'{selected_cause} percentages']
    y2_vals = dcauses[f'{selected_cause} minutes']

    fig = go.Figure()

    fig.add_trace(go.Bar(
        x=mm,
        y=y1_vals,
        name='%',
        marker_color='blue',
        yaxis='y1',
        width=0.4,
        offsetgroup=0
    ))

    fig.add_trace(go.Bar(
        x=mm,
        y=y2_vals,
        name='minutes',
        marker_color='orange',
        yaxis='y2',
        width=0.4,
        offsetgroup=1
    ))

    fig.update_layout(
        title=f"Results for {selected_cause}",
        barmode='group',
        xaxis=dict(title='Month'),
        yaxis=dict(title='Delayed Percentages (%) (Blue)',side='left'),
        yaxis2=dict(title='Delayed Minutes (Orange)',side='right',overlaying='y',showgrid=False)
    )

    return fig

from jupyter_dash import JupyterDash
ngrok.set_auth_token("2pgJb5n2nq2nHhAMVnCpKCTd6yc_2YeLGdPaQaRH2eUwKzDzb")
public_url = ngrok.connect(8050)
print(f"Dash app running at {public_url}")

app.run_server(debug=True, port=8050)

```
